```julia
using CSV
using DataFrames
using Plots
using Plots: mm
using Flux
using Statistics
import Random
```

# DNN for PK modeling

```julia
file = "./data/warfarin.csv"
df = CSV.read(file, DataFrame);
```

```julia
first(df,5)
```

```julia
# Forward fill missing values in column 'AMT'. As the AMT is the dose and only appear on the first colum for each patient
for i in 2:size(df, 1)
    if ismissing(df.AMT[i])
        df.AMT[i] = df.AMT[i-1]
    end
end

# Then delete that initial row for each patient
dropmissing!(df, :DV);
```

```julia
# Divide datasets: when DVID = 1, we have data for PK and when DVID = 2, we have data for PD
df_pk = df[(df.DVID .== 1), :]
df_pd = df[(df.DVID .== 2), :];
```

```julia
df_pk = select(df_pk, [:ID, :TIME, :WEIGHT, :AGE, :SEX, :AMT, :DV]);
```

```julia
first(df, 5)
```

```julia
Random.seed!(1)

patients = unique(df_pk.ID);

# Create an array to store scatter plots
scatter_plots = Plots.Plot[]

# Loop through each patient
for i in patients[rand(1:length(patients), 4)]
    df_plot = df_pk[df_pk.ID .== i, :]
   
    x_values = df_plot.TIME
    y_values = df_plot.DV

    push!(scatter_plots, scatter(x_values, 
                                 y_values,
                                 xlabel="Time",
                                 ylabel="DV",
                                 title="Patient $i",
                                 legend = false,
                                 markersize=2,
                                 markercolor=:blue,
                                 titlefont=font(12)
                                 )
            )
end

# Combine scatter plots into a grid layout
plot_grid = plot(scatter_plots..., layout=(2, 2), size=(800, 600), margin=5mm)

# Show the combined plot
display(plot_grid)
```

## Model

```julia
Random.seed!(1)
test_ids = patients[rand(1:length(patients), Int(length(patients)/4))]

train_df = filter(row -> !(row.ID in test_ids), df_pk)
select!(train_df, Not(:ID))

test_df = filter(row -> (row.ID in test_ids), df_pk)
select!(test_df, Not(:ID));
```

```julia
y_train = train_df[:, :DV]
X_train = Matrix(select(train_df, Not(:DV)));

y_test = test_df[:, :DV]
X_test = Matrix(select(test_df, Not(:DV)));
```

```julia
# Define the neural network architecture
model = Chain(
    Dense(5, 2, relu),   # Input layer: 5 input features, 16 neurons in the hidden layer, ReLU activation
    Dense(2, 1, softplus) # Output layer: 16 input neurons (from the hidden layer), 1 output neuron, Softplux activation so it is positive
)

# Define loss function (mean squared error)
loss(x, y) = Flux.mse(model(x), y)

# Define optimization algorithm (Stochastic Gradient Descent)
optimizer = Flux.Descent(0.01)
```

```julia
# Train the model
epochs = 1
for epoch in 1:epochs
    Flux.train!(loss, Flux.params(model), [(X_train, y_train)], optimizer)
    println("Epoch $epoch, Loss: $(loss(X_train, y_train))")
end
```

```julia
#Auxiliary functions for generating our data
function generate_real_data(n)
    x1 = rand(1,n) .- 0.5
    x2 = (x1 .* x1)*3 .+ randn(1,n)*0.1
    return vcat(x1,x2)
end
function generate_fake_data(n)
    θ  = 2*π*rand(1,n)
    r  = rand(1,n)/3
    x1 = @. r*cos(θ)
    x2 = @. r*sin(θ)+0.5
    return vcat(x1,x2)
end
# Creating our data
train_size = 5000
real = generate_real_data(train_size)
fake = generate_fake_data(train_size)
# Visualizing
scatter(real[1,1:500],real[2,1:500])
scatter!(fake[1,1:500],fake[2,1:500])
```

```julia
function NeuralNetwork()
    return Chain(
            Dense(2, 25,relu),
            Dense(25,1,x->σ.(x)) # Dense(25,1,sigmoid) 
            )
end
```

```julia
# Organizing the data in batches
X    = hcat(real,fake)
Y    = vcat(ones(train_size),zeros(train_size))
data = Flux.Data.DataLoader((X, Y'), batchsize=100,shuffle=true);
# Defining our model, optimization algorithm and loss function
m    = NeuralNetwork()
opt = Descent(0.05)
loss(x, y) = sum(Flux.Losses.binarycrossentropy(m(x), y))
```

```julia
ps = Flux.params(m)
epochs = 20
for i in 1:epochs
    Flux.train!(loss, ps, data, opt)
end
println(mean(m(real)), " ", mean(m(fake))) # Print model prediction
```

```julia
m    = NeuralNetwork()
function trainModel!(m,data;epochs=20)
    for epoch = 1:epochs
        for d in data
            gs = gradient(Flux.params(m)) do
                l = loss(d...)
            end
            Flux.update!(opt, Flux.params(m), gs)
        end
    end
    @show mean(m(real)),mean(m(fake))
end
trainModel!(m,data;epochs=20)
```
